```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print("Before append the array look like : ", arr)

arr = np.append(arr, [50])
print("After append the array look like : ", arr)
```

```python
import numpy as np

arr = np.array([[1, 2],[3, 4],[5, 6]])
print("Before Reshape array size look like : ", arr.size)
print("Array look like\n", arr)

arr = arr.reshape(2, 3)
print("After Reshape array size look like : ", arr.size)
print("Array look like \n", arr)
```
