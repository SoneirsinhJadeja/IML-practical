```python
import numpy as np 

ar = np.arange(6)
ar = np.random.randint(10, size=5)
# First array is create
ar2 = np.arange(6)
ar2 = np.random.randint(10, size=5)
# Second array is create

add = np.add(ar, ar2) # addtion of two numpy arr is 
sub = np.subtract(ar, ar2) # addtion of two numpy arr is 
mul = np.multiply(ar, ar2) # addtion of two numpy arr is 
div = np.divide(ar, ar2) # addtion of two numpy arr is 

print("\n First array \n", ar)
print("\n Second array \n", ar2)
print("\n Addition of two arr is :\n", add)
print("\n Subtraction of two arr is :\n", sub)
print("\n Multiplications of two arr is :\n", mul)
print("\n Division of two arr is :\n", div)
```
