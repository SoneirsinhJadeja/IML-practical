```python
import pandas as pd

dic = {
    "name" : ["Cat", "Ball", "Apple"],
    "color" : ["grey", "bule", "red"]
}

arr = pd.DataFrame(dic)

val  =  arr.sort_values(by=["name"], axis=0, ascending=True)

print(val)
```
